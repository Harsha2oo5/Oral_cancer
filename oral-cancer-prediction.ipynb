```python
# 1. Imports
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model

# 2. Directory setup
base_dir = '/kaggle/input/oral-cancer-dataset2/OralCancer'  # Point to folder containing 'cancer' and 'non-cancer'

# 3. Image data generators (with augmentation)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='training',
    shuffle=True
)

val_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='validation',
    shuffle=True
)

# 4. Load pre-trained VGG16 model (without top classification layer)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# 5. Freeze base model layers (only train new layers initially)
base_model.trainable = False

# 6. Add custom classification head
x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

# 7. Compile model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 8. Train model (Phase 1: frozen base)
print("Training Phase 1: Base model frozen")
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)

# 9. Unfreeze base model and fine-tune (optional but recommended)
base_model.trainable = True
model.compile(
    optimizer=Adam(learning_rate=0.00001),  # Lower learning rate for fine-tuning
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("Training Phase 2: Fine-tuning entire model")
history_fine = model.fit(
    train_generator,
    epochs=5,
    validation_data=val_generator
)

# 10. Performance visualization
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'] + history_fine.history['accuracy'], label='Train acc')
plt.plot(history.history['val_accuracy'] + history_fine.history['val_accuracy'], label='Val acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'] + history_fine.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'] + history_fine.history['val_loss'], label='Val loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')
plt.show()

# 11. Save trained model
model.save('/kaggle/working/oral_cancer_vgg16_model.h5')
print("Model saved at /kaggle/working/oral_cancer_vgg16_model.h5")
```

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# Load model
model = load_model('/kaggle/working/oral_cancer_vgg16_model.h5')

# Check class mapping (run this once to verify)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
temp_gen = ImageDataGenerator(rescale=1./255)
temp = temp_gen.flow_from_directory(
    '/kaggle/input/oral-cancer-dataset2/OralCancer',
    target_size=(128, 128),
    batch_size=1,
    class_mode='binary'
)
print("Class indices:", temp.class_indices)
# Should output: {'cancer': 0, 'non-cancer': 1}

# Create reverse mapping
class_labels = {v: k for k, v in temp.class_indices.items()}
print("Class labels:", class_labels)

# Test on NON-CANCER images
print("="*50)
print("Testing on NON-CANCER images (should predict 'non-cancer')")
print("="*50)

non_cancer_folder = '/kaggle/input/oral-cancer-dataset2/OralCancer/non-cancer'
files = [f for f in os.listdir(non_cancer_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

correct = 0
total = 0

for img_file in random.sample(files, min(5, len(files))):
    img_path = os.path.join(non_cancer_folder, img_file)
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    pred_score = model.predict(img_array, verbose=0)[0][0]
    
    # Use class mapping
    pred_class_idx = 1 if pred_score > 0.5 else 0
    pred_class = class_labels[pred_class_idx]
    
    if pred_class == 'non-cancer':
        correct += 1
    total += 1
    
    print(f"{img_file}: {pred_class} (score: {pred_score:.4f})")
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"TRUE: non-cancer | Predicted: {pred_class} ({pred_score:.2f})")
    plt.show()

print(f"Accuracy on non-cancer: {correct}/{total}")

# Test on CANCER images
print("\n" + "="*50)
print("Testing on CANCER images (should predict 'cancer')")
print("="*50)

cancer_folder = '/kaggle/input/oral-cancer-dataset2/OralCancer/cancer'
files = [f for f in os.listdir(cancer_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

correct = 0
total = 0

for img_file in random.sample(files, min(5, len(files))):
    img_path = os.path.join(cancer_folder, img_file)
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    pred_score = model.predict(img_array, verbose=0)[0][0]
    
    pred_class_idx = 1 if pred_score > 0.5 else 0
    pred_class = class_labels[pred_class_idx]
    
    if pred_class == 'cancer':
        correct += 1
    total += 1
    
    print(f"{img_file}: {pred_class} (score: {pred_score:.4f})")
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"TRUE: cancer | Predicted: {pred_class} ({pred_score:.2f})")
    plt.show()

print(f"Accuracy on cancer: {correct}/{total}")
```
